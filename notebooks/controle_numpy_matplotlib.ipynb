# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
import numpy as np
import matplotlib.pyplot as plt
data = np.genfromtxt('Iris.csv' , delimiter=',' , skip_header=1, dtype= object)
print(data.shape)
```

# Partie D — Visualisation avec Matplotlib (Plots Simples)

### D17. Line plot
Line plot des longueurs des sépales pour les 50 premières fleurs.


```python
import matplotlib.pyplot as plt

plt.plot(features[:50, 0], marker='o')
plt.title("Longueur des sépales – 50 premières fleurs")
plt.xlabel("Index")
plt.ylabel("Sepal length (cm)")
plt.grid(True)
plt.show()
```

### D18. Scatter plot
Scatter plot : longueur vs largeur des pétales.


```python
plt.scatter(features[:, 2], features[:, 3], c='tab:green', alpha=0.7)
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Longueur vs largeur des pétales')
plt.grid(True)
plt.show()
```

### D19. Bar chart
Bar chart des moyennes des quatre caractéristiques.


```python
moyennes = np.mean(features, axis=0)
labels = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']

plt.bar(labels, moyennes, color=['skyblue','orange','lightgreen','coral'])
plt.ylabel('Moyenne (cm)')
plt.title('Moyenne des quatre caractéristiques')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### D20. Histogram
Histogramme de la largeur des sépales (10 bins).


```python
plt.hist(features[:, 1], bins=10, color='steelblue', edgecolor='black')
plt.xlabel('Sepal width (cm)')
plt.ylabel('Fréquence')
plt.title('Histogramme de la largeur des sépales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### D21. Boxplot par espèce
Boxplot des longueurs des pétales par espèce.


```python
plt.boxplot(features[:, 2].reshape(3, 50).T,
            labels=['setosa', 'versicolor', 'virginica'])
plt.ylabel('Petal length (cm)')
plt.title('Boxplot des longueurs des pétales par espèce')
plt.show()
```

# Partie E — Visualisations Avancées

### E22. Subplots comparatifs
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python
plt.figure(figsize=(9,4))

plt.subplot(1,2,1)
plt.scatter(features[:50,2], features[:50,3])
plt.title("Setosa")

plt.subplot(1,2,2)
plt.scatter(features[50:100,2], features[50:100,3])
plt.title("Versicolor")

plt.tight_layout()
plt.show()
```

### E23. Grille 2x2
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python
fig, ax = plt.subplots(2,2, figsize=(7,5))
for i in range(4):
    plt.sca(ax.flat[i])
    plt.hist(features[:,i])
plt.tight_layout(); plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
for i, name in enumerate(['Setosa','Versicolor','Virginica']):
    m = features[50*i:50*(i+1)].mean(0)
    s = features[50*i:50*(i+1)].std(0)
    plt.errorbar(range(4), m, s, label=name, marker='o')

plt.xticks(range(4), ['SL','SW','PL','PW']);
plt.legend();
plt.show()
```

### E25. Scatter coloré par espèce
Scatter plot avec couleurs par espèce.


```python
plt.scatter(features[:, 2], features[:, 3], c=features[:,2]//2+1)
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Longueur vs largeur des pétales')
plt.grid(True)
plt.show()
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
plt.plot(np.cumsum(features[:,3].reshape(3,50).T, 0))
plt.legend(['Set','Ver','Vir']);
plt.ylabel('Cumsum');
plt.show()
```

# Partie F — Mini-Analyse Finale

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
corr = np.corrcoef(features, rowvar=False)
print(corr)
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
plt.scatter(*np.where(np.abs(corr) < 1), c = corr[np.where(np.abs(corr) < 1)] )
plt.colorbar(label='correlation')
plt.xlabel('feature i');
plt.ylabel('feature j');
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
most_var = np.argmax([features[i*50:(i+1)*50].var() for i in range(3)])
plt.figure(figsize=(6,2))
for k in range(4):
    plt.subplot(1,4,k+1)
    plt.hist(features[most_var*50:(most_var+1)*50,k])
plt.suptitle(['Setosa','Versicolor','Virginica'][most_var])
plt.tight_layout();
plt.show()
```
